```python
import os
import pymorphy2
from nltk.tokenize import sent_tokenize
```

```python
punct = '”“".,«»\\/*!:;—()\'-%`.?―[]'
morph = pymorphy2.MorphAnalyzer()
#result = open('corpora.txt', 'a', encoding='utf-8')
files = os.listdir('.')
for name in files:
    if not name.endswith('.py') and not name.endswith('.DS_Store') and not name.endswith('.ipynb'):
        print(name)
        f = open(name, 'r', encoding='utf')
        for line in f:
            if line[0] not in '1234567890':
                line = line.strip().lower()
                line1 = sent_tokenize(line, language='english')
                for sentence in line1:
                    print(sentence)
                    new_sentence = ''
                    sentence = sentence.split(' ')
                    n = 1
                    for word in sentence:
                        word = word.strip(punct).lstrip(punct)
                        word = word.strip(punct)
                        try:
                            new_word = morph.parse(word)[0].normal_form + '_' + morph.parse(word)[0].tag.POS
                        except TypeError:
                            new_word = word
                        finally:
                            if n == 1:
                                new_sentence += new_word
                                n = 2
                            else:
                                new_sentence += ' ' + new_word
                    print(new_sentence)
                    print('!!!')
        #f.close()
    break
```

```python
punct = '”“".,«»\\/*!:;—()\'-%`.?―[]'
morph = pymorphy2.MorphAnalyzer()
corpus = ''
corp_len = 0
#result = open('corpora.txt', 'w', encoding='utf-8')
files = os.listdir('.')
for name in files:
    if name.endswith('txt'):
        print(name)
        f = open(name, 'r', encoding='utf')
        for line in f:
            if line[0] not in '1234567890':
                line = line.strip().lower()
                line1 = sent_tokenize(str(line), language='english')
                for sentence in line1:
                    new_sentence = ''
                    sentence = sentence.split()
                    n = 1
                    for word in sentence:
                        corp_len += 1
                        word = word.strip(punct).lstrip(punct)
                        word = word.strip(punct)
                        word = word.strip(punct)
                        try:
                            new_word = morph.parse(word)[0].normal_form + '_' + morph.parse(word)[0].tag.POS
                        except TypeError:
                            new_word = word
                        finally:
                            if n == 1:
                                new_sentence += new_word
                                n = 2
                            else:
                                new_sentence += ' ' + new_word
                    corpus += new_sentence + '\n'
        f.close()
#result.write(corpus)
#result.close()
#print('Done')
print(corp_len)
```
